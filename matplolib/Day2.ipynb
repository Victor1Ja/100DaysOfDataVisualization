# Line chart


```python
import random as rd
import matplotlib.pyplot as plt

def random_array(len:int, ini = 1, end = 10):
    return [rd.randint(ini, end) for x1  in range(len)]

def parabole(xPoints):
    return [(x**2) for x in xPoints]

amount_of_points = 30

xPoints = range(amount_of_points)
yPoints = parabole(xPoints)
y1Points = parabole([x*0.9 for x in range(amount_of_points)])

figure = plt.figure()
plt.plot(range(3,amount_of_points+3), yPoints,'-o')
plt.plot( yPoints, '-o', y1Points, '-o')
plt.legend(['Linear', 'Quadratic', 'Quadratic *0.9'])
plt.show()
```

```python
plt.figure()
plt.plot(xPoints, yPoints, '-o')
plt.fill_between(xPoints, yPoints, y1Points, color='blue', alpha=0.1)
plt.show()
```

# Pie Chart


```python
import numpy as np
cars = np.array(['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES'])
data = np.array([23, 17, 35, 29, 12, 41])
total_data = data.sum()
new_data = data/total_data*100
print(new_data, new_data.sum())

# Note pie show the percent of the total LOL it is not necessary to give it percents
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)
fig = plt.figure(figsize =(10, 7))
plt.pie(new_data, labels = cars)
plt.show()
```

## A more eatable Pie


```python

# Creating dataset
ingredients = ['Milk', 'Cheese', 'Candies',
	    'Sugar', 'Flour', 'Chocolate']

data = [23, 17, 35, 29, 12, 41]


# The explode is the effect of distance from the center
explode = (0.5, 0.4, 0.3, 0.2, 0.1, 0.0)

# Creating color parameters
colors = ( "orange", "cyan", "brown",
		"grey", "indigo", "beige")

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

# autopct is for render the text inside the piece of pie
def func(pct, allvalues):
	absolute = int(pct / 100.*np.sum(allvalues))
	return "{:.2f}%\n({:d} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
								autopct = lambda pct: func(pct, data),
								explode = explode,
								labels = ingredients,
								shadow = True,
								colors = colors,
								startangle = 90,
								wedgeprops = wp,
								textprops = dict(color ="magenta"))

# Adding legend
ax.legend(wedges, ingredients,
		title ="Ingredients",
		loc ="center left",
		bbox_to_anchor =(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")

# show plot
plt.show()
```

## Pie inside a Pie

Yummy ;-)


```python
# Creating dataset
ingredients = ['Milk', 'Cheese', 'Candies',
	    'Sugar', 'Flour', 'Chocolate']

data = np.array([[23, 16], [17, 23],
                 [35, 11], [29, 33],
                 [12, 27], [41, 42]])

size = data.shape[0]

# normalizing data to 2 pi
norm = data / np.sum(data)*2 * np.pi


cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(6)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9,
                              10, 12, 13, 15,
                              17, 18, 20 ]))
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7),
                       subplot_kw = dict(polar = True))
 
print(norm.flatten())
# obtaining ordinates of bar edges
left = np.cumsum(np.append(0,
                           norm.flatten()[:-1])).reshape(data.shape)
# This is more than a pie a BarChart that uses polar coordinates, 
# the first plot it sum the values of the axis 1 and 
ax.bar(x = left[:, 0],
       width = norm.sum(axis = 1),
       bottom = 1-size,
       height = size,
       color = outer_colors,
       edgecolor ='w',
       linewidth = 1,
       align ="edge")
 
ax.bar(x = left.flatten(),
       width = norm.flatten(),
       bottom = 1 - (2 * size),
       height = size,
       color = inner_colors,
       edgecolor ='w',
       linewidth = 1,
       align ="edge")
 
ax.set(title ="Nested pie chart")
ax.set_axis_off()
 
# show plot
plt.show()
```
