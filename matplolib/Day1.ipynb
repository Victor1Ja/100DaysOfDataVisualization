# Scatter plot


```python
import random as rd
import matplotlib.pyplot as plt

def random_array(len:int, ini = 1, end = 10):
    return [rd.randint(ini, end) for x1  in range(len)]

def parabole(xPoints):
    return [(x**2) for x in xPoints]

amount_of_points = 100


x_axis = random_array(amount_of_points, 1, 50)
y_axis = random_array(amount_of_points, 1, 50)
# x = range(-50,50)
# y = parabole(x)

plt_figure_id = 1

figure = plt.figure(plt_figure_id)

plt.scatter(x_axis,y_axis, c='blue')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```
